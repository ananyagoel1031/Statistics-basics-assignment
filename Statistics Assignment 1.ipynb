```python
#Q6
import numpy as np

list_x=[10,20,30,40,50]
list_y=[15,25,35,45,60]
```

```python
# Compute covariance matrix
cov_matrix = np.cov(list_x, list_y)

# Extract covariance value
covariance = cov_matrix[0][1]
print(f"Covariance: {covariance}")
```

```python
# Compute correlation coefficient matrix
corr_matrix = np.corrcoef(list_x, list_y)

# Extract correlation coefficient value
correlation = corr_matrix[0][1]
print(f"Correlation Coefficient: {correlation}")
```

```python
#Q7

import matplotlib.pyplot as plt

# Data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create the boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add labels and title
plt.title("Boxplot of Data")
plt.xlabel("Values")

# Display the plot
plt.show()
```

```python
#Q8

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute correlation coefficient matrix
corr_matrix = np.corrcoef(advertising_spend, daily_sales)

# Extract correlation coefficient value
correlation = corr_matrix[0][1]
print(f"Correlation Coefficient: {correlation}")
```
